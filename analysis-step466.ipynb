```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
# I chose to analyze the 'price' variable as it's the easiest one to work with in the dataset
 
print("UNIVARIATE ANALYSIS - PRICE")
print("=" * 50)
 
price = diamonds['price']
 
# stats
print(f"Mean: {price.mean()}")
print(f"Median: {price.median()}")
print(f"Std Dev: {price.std()}")
print(f"Min: {price.min()}")
print(f"Max: {price.max()}")
```

```python
# data visuals
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(price, bins=30, color='blue', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price')
 
plt.subplot(1, 2, 2)
plt.boxplot(price)
plt.ylabel('Price')
plt.title('Boxplot of Price')
plt.tight_layout()
plt.show()
```

```python

# I chose to analyze the relationship between 'carat' and 'price'
 
print("\nMULTIVARIATE ANALYSIS - CARAT vs PRICE")
print("=" * 50)
 
# correlation between the two vars
correlation = diamonds['carat'].corr(diamonds['price'])
print(f"Correlation between carat and price: {correlation:.4f}")
 
# data visuals
plt.figure(figsize=(10, 5))
plt.scatter(diamonds['carat'], diamonds['price'], alpha=0.5)
plt.xlabel('Carat Weight')
plt.ylabel('Price')
plt.title('Price vs Carat Weight')
plt.grid(True)
plt.show()
 
# show average price
print("\nAverage price by cut quality:")
print(diamonds.groupby('cut')['price'].mean())
```
