```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data set
countries = pd.read_csv("countries.csv")
```

```python
# display first 5 rows of the dataset
countries.head()
```

```python
# filtering the dataset to only include rows from 1952 to compare life expectancy for Asia and Africa in 1952
data_1952 = countries[countries["year"] == 1952]
```

```python
# dataframe for Asian countries1952
asia_1952 = data_1952[data_1952["continent"] == "Asia"]

# dataframe for African countries 1952
africa_1952 = data_1952[data_1952["continent"] == "Africa"]
```

```python
# 2 subplots side-by-side to compare the life expectancy distribution of Asia and Africa

plt.figure(figsize=(12, 5))

# asia subplot
plt.subplot(1, 2, 1)
plt.hist(asia_1952["lifeExpectancy"], bins=10)
plt.title("Life Expectancy in Asia, 1952")
plt.xlabel("Life Expectancy")
plt.ylabel("Number of Countries")

# africa subplot
plt.subplot(1, 2, 2)
plt.hist(africa_1952["lifeExpectancy"], bins=10)
plt.title("Life Expectancy in Africa, 1952")
plt.xlabel("Life Expectancy")
plt.ylabel("Number of Countries")

# spacing
plt.tight_layout()

# display
plt.show()
```
