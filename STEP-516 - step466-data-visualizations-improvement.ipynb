```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# setting the theme for the visuals
sns.set_theme(style='whitegrid', context='notebook')
plt.rcParams['figure.dpi'] = 120
plt.rcParams['axes.titleweight'] = 'bold'
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
 
print("UNIVARIATE ANALYSIS - PRICE")
 
price = diamonds['price']
 
# stats
print(f"Mean: {price.mean()}")
print(f"Median: {price.median()}")
print(f"Std Dev: {price.std()}")
print(f"Min: {price.min()}")
print(f"Max: {price.max()}")
```

```python
# data visuals
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

sns.histplot(price, bins=35, kde=True, color='#4C78A8', edgecolor='white', alpha=0.9, ax=axes[0])
axes[0].set_xlabel('Price')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Distribution of Diamond Prices')

sns.boxplot(y=price, color='#F58518', width=0.35, fliersize=2, ax=axes[1])
axes[1].set_ylabel('Price')
axes[1].set_title('Price Spread and Outliers')

sns.despine()
fig.tight_layout()
plt.show()
```

```python
 
print("\nMULTIVARIATE ANALYSIS - CARAT vs PRICE")
print("=" * 50)
 
# correlation between the two vars
correlation = diamonds['carat'].corr(diamonds['price'])
print(f"Correlation between carat and price: {correlation:.4f}")
 
# data visuals
plt.figure(figsize=(10.5, 5.5))

sns.scatterplot(
    data=diamonds,
    x='carat',
    y='price',
    alpha=0.22,
    s=18,
    color='#54A24B',
    edgecolor=None
)

# adding trendline for better visuals and readability
sns.regplot(
    data=diamonds,
    x='carat',
    y='price',
    scatter=False,
    color='#E45756',
    line_kws={'linewidth': 2}
)

plt.xlabel('Carat Weight')
plt.ylabel('Price')
plt.title('Price vs Carat Weight (with Trendline)')
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
 
# ave price
print("\nAverage price by cut quality:")
print(diamonds.groupby('cut')['price'].mean())
```
